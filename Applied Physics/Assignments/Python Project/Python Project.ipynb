# Question 01 (Vectors)
### What are (a) the x component and (b) the y component of avector in the xy plane if its direction is 250Â° counterclockwise from the positive direction of the x axis and its magnitude is 7.3 m?

```python
import math
import matplotlib.pyplot as plt
import numpy as np
magnitude=7.3
angle=math.radians(250)
xcomp= magnitude*math.cos(angle)
ycomp= magnitude*math.sin(angle)
print("The magnitude of x-component is",xcomp,"and the magnitude of the y-component is",ycomp)
X = np.array(0)
Y= np.array(0)
U = np.array(xcomp)
V = np.array(ycomp)
fig, ax = plt.subplots()
a = ax.quiver(X, Y, U, V,color = 'r', units='xy' ,scale=1.5)

A=np.array(xcomp)
B=np.array(0)
Ax=ax.quiver(A,B,color = 'b', units='xy' ,scale=1.5)
C=np.array(0)
D=np.array(ycomp)
Ax=ax.quiver(C,D,color = 'b', units='xy' ,scale=1.5)

plt.grid()

ax.set_aspect('equal')

plt.xlim(-5,5)
plt.ylim(-5,5)
plt.plot()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Vector and its components representation')
```

# Question 05 (Uniform Circular Motion)
## An Earth satellite moves in a circular orbit 640 km (uniform circular motion) above Earthâs surface with a period of 98.0 min. What are (a) the speed and (b) the magnitude of the centripetal acceleration of the satellite?

```python
import matplotlib.pyplot as plt
import numpy as np
import math
radius_of_earth =6.37e6
Total_radius=radius_of_earth+640000
Time_period=98*60
speed=(2*math.pi*Total_radius)/(Time_period)
cent_acc=speed**2/radius_of_earth
print("The speed of the satellite is ",speed,"m/s\nAnd the magnitude of the centripetal acceleration is ",cent_acc,"m/s^2")
circle2 = plt.Circle((0, 0),Total_radius,fc="none",ec='red')
fig, ax = plt.subplots()
ax.add_artist(circle2)
ax.set_aspect('equal')
a= np.array(Total_radius)
b= np.array(0)
Radius = ax.quiver(a,b,color = 'r', units='xy' ,scale=1)
c= np.array(Total_radius)
d= np.array(0)
e=np.array(0)
f=np.array(Total_radius)
ax.arrow(c,d,e,f,color='blue')
Centripetal_acceleration= ax.quiver(c,d,e,f,color = 'b', units='xy' ,scale=1)
plt.grid()
plt.xlim(-1e7,1e7)
plt.ylim(-1e7,1e7)
plt.title("Orbit and radius of the satellite\nDirection of velocity(with blue)",fontsize=10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
fig.savefig('plotcircles.png')
```
