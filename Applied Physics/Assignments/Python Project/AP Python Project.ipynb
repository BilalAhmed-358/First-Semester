# Question 07 (Newton's laws)
## Only two horizontal forces act on a 3.0 kg body that can move over a frictionless floor. One force is 9.0 N, acting due east, and the other is 8.0 N, acting 62 degree north of west.What is the magnitude of the bodyâs acceleration?

```python
import matplotlib.pyplot as plt
import numpy as np
import math
F1=9
theta1=0
angle1=np.radians(theta1)
F2=8
theta2 =180-62
angle2=np.radians(theta2)
F3x=F1*np.cos(angle1)+F2*np.cos(angle2)
F3y=F1*np.sin(angle1)+F2*np.sin(angle2)
angle3=math.atan(F3y/F3x)
F3=np.sqrt((F3x)**2+(F3y)**2)
mass=3
acceleration=(F3)/mass
print("The magnitude of the body's acceleration is ",angle3,"m/s^2")
a= np.array(F1*np.cos(angle1))
b= np.array(F1*np.sin(angle1))
c = np.array(F2*np.cos(angle2))
d = np.array(F2*np.sin(angle2))
e=np.array(acceleration*np.cos(angle3))
f=np.array(acceleration*np.sin(angle3))
fig, ax = plt.subplots()
F_1 = ax.quiver(a,b,color = 'r', units='xy' ,scale=1)
F_2 = ax.quiver(c,d,color = 'b', units='xy' ,scale=1)
acceleration = ax.quiver(e,f,color = 'purple', units='xy' ,scale=1)
plt.grid()
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.plot()
plt.xlabel('x-axis\n\nFree body diagram',fontsize=12)
plt.ylabel('y-axis',fontsize=12)
plt.title("Free body diagram")
plt.title('Vector representation of Forces acting on the body\nF1 in red\nF2 in blue\ndirection of acceleration in purple')
```

# Question 08 (Simple Harmonic Motion)
## An object undergoing simple harmonic motion takes 0.25 s to travel from one point of zero velocity to the next such point. The distance between those points is 36 cm. Calculate the (a) period, (b) frequency, and (c) amplitude of the motion.

```python
import math
import matplotlib.pyplot as plt
import numpy as np
time= 0.25
distance=36
Time_period=2*time
frequency=1/Time_period
amplitude=distance/2
print("The time period is",Time_period,"s\nThe frequency is",frequency,"Hz\nThe amplitude is",amplitude,"cm")
time=np.arange(0.001,20,.1)
w=(2*3.142)/time
phs=0
def disp(amplitude,w,time,phs):
    disp=amplitude*np.cos(w*time+phs)
    return disp
plt.title("This is the graph of displacement",fontsize='12')
plt.plot(time,disp(12,15,time,6))
plt.grid()
plt.show()
def vel(amp,w,time,phs):
    vel=amplitude*w*(np.sin(w*time+phs))
    return vel
plt.title("This is the graph of velocity",fontsize='12')
plt.plot(time,vel(6,5,time,.1))
plt.grid()
plt.show()
def acc(amplitude,w,time,phs):
    acc=amplitude*(w**2)*(np.cos(w*time+phs))
    return acc
plt.title("This is the graph of acceleration",fontsize='12')
plt.plot(time, acc(20,30,time,1))
plt.grid()
plt.show()
```
