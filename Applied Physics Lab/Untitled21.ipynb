```python
import math
import numpy
def Vector_Operation (ax,ay,bx,by):
    dotProduct = ax*bx + ay*by
    productofmagnitude = math.sqrt( ax*ax + ay*ay)*math.sqrt(bx*bx + by*by) 
    angle=math.acos(dotProduct/productofmagnitude)
    vectorProduct=math.sqrt( ax*ax + ay*ay)*math.sqrt(bx*bx + by*by)*math.sin(angle)
    angle=numpy.degrees(angle)
    return dotProduct,vectorProduct,angle
Vector_Operation (3,4,5,6)
```

```python
import matplotlib.pyplot as plt
import numpy as np
time= np.arange(0,2*np.pi)
amplitude=len("BilalAhmed")
w=np.pi*amplitude
phase=np.pi/2

def displacement(amplitude,w,t,phase):
    displacement=amplitude*np.cos(w*t+phase)
    return displacement
def velocity(amplitude,w,t,phase):
    velocity=amplitude*w*np.sin(w*t+phase)
    return velocity
def acceleration(amplitude,w,t,phase):
    acceleration=amplitude*(w**2)*np.cos(w*t+phase)
    return acceleration
plt.plot(displacement(12,24,time,np.pi),time)
plt.show()
plt.plot(velocity(12,11,time,np.pi),time)
plt.show()
plt.plot(acceleration(12,24,time,np.pi),time)
plt.show()
    
```
